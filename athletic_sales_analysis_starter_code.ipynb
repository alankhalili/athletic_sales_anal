### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
def  comma(x):
    return '{:,.0f}'.format(x)
```

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
df1=df.groupby(['region','state','city'])[['units_sold']].sum()
df1=df1.rename(columns={'units_sold':'Total_Products_Sold'})
df1=df1.sort_values(['Total_Products_Sold'], ascending=False)
#df1=df1['Total_Products_Sold'].apply(comma)

# Show the top 5 results.
df1.head(5)                           
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
df2=pd.pivot_table(df,values='units_sold',index=['region','state','city'],aggfunc='sum')


# Rename the "units_sold" column to "Total_Products_Sold"
df2=df2.rename(columns={'units_sold':'Total_Products_Sold'})

# Show the top 5 results.
df2=df2.sort_values(['Total_Products_Sold'], ascending=False)
df2.head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
df_waf=df.loc[df['product']=="Women's Athletic Footwear"]
df_waf.head(5)
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
df7=df_waf.groupby(['retailer','region','state','city'])[['units_sold']].sum()


# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
df7=df7.rename(columns={'units_sold':"Women's Footwear Units Sold"})
# Show the top 5 results.
df7=df7.sort_values(["Women's Footwear Units Sold"], ascending=False)
df7.head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
df8=pd.pivot_table(df_waf,values='units_sold',index=['retailer','region','state','city'],aggfunc='sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
df8=df8.rename(columns={'units_sold':'Womens_Footwear_Units_Sold'})
df8=df8.sort_values(['Womens_Footwear_Units_Sold'], ascending=False)

# Show the top 5 results.
df8.head()
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
df9=pd.pivot_table(df_waf,values='total_sales',index=['invoice_date'],aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
df9=df9.rename(columns={'total_sales':'Total Sales'})

# Show the table.
df9
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
df10=df9.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
df10=df10.sort_values(['Total Sales'], ascending=False)
df10
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
df11=df9.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
df11=df11.sort_values(['Total Sales'], ascending=False)
df11
```
